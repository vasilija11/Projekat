# Elektrane u svijetu

```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline 

import pandas as pd
import numpy as np
df=pd.read_csv('global_power_plant_database.csv')

lista_drzava = list(df.country_long)
sirina = list(df.latitude)
duzina = list(df.longitude)

# Make a data frame with the GPS of a few cities:
data = pd.DataFrame({
'lat': sirina,
'lon': duzina,
'name': lista_drzava
})

 
# A basic map
m=Basemap(llcrnrlon=-160, llcrnrlat=-75,urcrnrlon=160,urcrnrlat=80)
m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m.fillcontinents(color='grey', alpha=0.7, lake_color='grey')
m.drawcoastlines(linewidth=0.1, color="white")
 
# Add a marker per city of the data frame!
m.plot(data['lat'], data['lon'], linestyle='none', marker="o", markersize=3,alpha=0.4,c="red", markeredgecolor="yellow",markeredgewidth=1)

```
